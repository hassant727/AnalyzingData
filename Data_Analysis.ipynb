### Import Libraries

```python
# Importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### Cleaning The Data

```python
# Reading in the data
df = pd.read_csv("hospital_data.csv")
df.head()
```

```python
# Get rid of whitespaces
df.columns = df.columns.str.strip()
df.columns
```

```python
# Look at the diffierent data types to see what transformations you might have to do
df.dtypes
```

```python
# Change datatype of columns from object to datetime
df['Date'] = pd.to_datetime(df['Date'])
df['Post-Consultation Time'] = pd.to_datetime(df['Post-Consultation Time'])
df['Entry Time'] = pd.to_datetime(df['Entry Time'])
df['Completetion Time'] = pd.to_datetime(df['Completion Time'])
df.dtypes
```

```python
# Function to clean text
def purge(p):
    # remove the dollar sign and replace with nothing
    p = p.replace('$', '')
    # remove commas and replace with nothing
    p = p.replace(',', '')
    # remove dash and repalce with nothing
    p = p.replace('-', '')
    # remove space and repacle with nothing
    p = p.replace(' ', '')
    # change to float numeric value
    p = pd.to_numeric(p)
    return p

df['Medication Revenue'] = df['Medication Revenue'].apply(purge)
df.head()
```

```python
# Apply function to required columns
df['Consultation Revenue'] = df['Consultation Revenue'].apply(purge)
df['Lab Cost'] = df['Lab Cost'].apply(purge)
df.dtypes


    
```

### Doing Analysis

```python
# Create new column - Total Revenue
df['Total Revenue'] = df['Medication Revenue'] + df['Consultation Revenue'] + df['Lab Cost']
df.head()
```

```python
# Get rid of NaN values in table
df = df.fillna(0)
df.head()
```

### Business Analysis

# Q: What time has the most visitors


```python
# Use series.dt to aaccess value sof the series as datetime, will retunr the hour of datetime
df['Entry Hour'] = df['Entry Time'].dt.hour
df.head()
```

```python
# Group by the hour to get all the different visitors
pivot1 = df.groupby(df['Entry Hour'])['Patient ID'].count()
pivot1
```

### Visulaizations

```python
plt.style.use('ggplot')
pivot1 = df.groupby(df['Entry Hour'])['Patient ID'].count().plot()
```

# Q: Does the total revenue correlate with the total vistors.

```python
pivot2 = df.groupby(df['Entry Hour'])['Total Revenue'].sum().plot()
```

```python
# Dispaly both graphs on same table
pivot3 = df.groupby(df['Entry Hour']).sum()
pivot4 = df.groupby(df['Entry Hour']).count()
merge = pivot3.merge(pivot4, on='Entry Hour')
merge.head()
```

```python
# Display table on for both graphs on one graph
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.bar(merge.index, merge['Total Revenue_x'],color='blue')
ax1.plot(merge.index, merge['Patient ID'])
```
